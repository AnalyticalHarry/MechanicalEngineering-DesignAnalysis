#### Import Libraries

```python
import sympy as sp
```

#### Example

A rectangular beam 60mm wide and 150mm deep is simply supported over a span of 4m. if the beam is subjected to a central point load of 10kN, find the maximum bending stress induced in the beam section.

#### 1. Rectangular box dimension

#### Given: 

length = 4000mm

width = 60mm

height = 150mm

```python
print("\nINPUT:")
l = float(input("Length of rectangular beam: "))
w = float(input("Width of rectangular beam: "))
h = float(input("Height of rectangular beam: "))
def rectangle_beam(l,w,h):
        surface_area = (2*l*w)+(2*l*h)+(2*w*h)
        volume = l*w*h
        c = sp.sqrt((l*l)+(w*w))
        d = sp.sqrt((l*l)+(w*w)+(h*h))
        y = h/2
        return surface_area, volume,c, d, y
rectangle_beam(l, w, h)
rectangle_beam = rectangle_beam(l, w, h)
print("\nOUTPUT:")
print(f"Surface Area: {round(rectangle_beam[0], 2)} milimeter square")
print(f"Volume: {round(rectangle_beam[1], 2)} milimeter cube")
print(f"Diagonal base of box: {round(rectangle_beam[2], 2)} milimeter") #Hypotenuse
print(f"Diagonal of entire box: {round(rectangle_beam[3], 2)} milimeter")
print(f"y: {round(rectangle_beam[4], 2)} milimeter")

y = rectangle_beam[4]
```

#### 2. Bending Moment

```python
p = float(input("Enter point load (kN/m): "))#load
def bending_moment(p,l):
    M = (p*l*1000)/4
    return M
bending_moment(p,l)
result_bending = bending_moment(p,l)
bending_moment = [result_bending]
print(f"Bending Moment: {bending_moment[0]} Nmm")

M = bending_moment[0]
```

#### 3. Momemnt of Inertia

```python
def moment_of_inertia(w,h):
    I = ((w)*((h)**3))/12
    return I
moment_of_inertia(w,h)
result_moment_of_inertia = moment_of_inertia(w,h)
moment_of_inertia = [result_moment_of_inertia]
print(f"Moment of inertia: {moment_of_inertia[0]} mm^4")

I = moment_of_inertia[0]
```

#### 4. Maximum bending stress induced in the beam section

```python
stress_max = (M/I)*y
print(f"Maimum bending stress: {round(stress_max,2)} MPa")
```
