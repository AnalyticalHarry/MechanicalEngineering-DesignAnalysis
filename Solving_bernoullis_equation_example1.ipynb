#### Algebraically solving Bernoulli’s equation for P2 - Using OOP method

```python
class Bernoullis_equation:
        def __init__(self, p1:float,rho:float,v1:float,v2:float, g:float,h1:float,h2:float):
            self.p1 = p1 #12300 pascal
            self.rho = rho #1090 Kg/m3
            self.v1 = v1 #3 m/s
            self.v2 = v2 #0.750 m/2
            self.g = g #9.81 m/s^2
            self.h1 = h1 # zero
            self.h2 = h2 #1.20 meter
            
        def Bernoullis_equation_p2(self):
            p2 = (self.p1)+(1/2*self.rho*self.v1**2)+self.rho*self.g*self.h1 - (1/2*self.rho*self.v2**2)-(self.rho*self.g*self.h2)
            return p2
        
p2 = Bernoullis_equation(12300,1090,3,0.750,9.81, 0,1.20)
print(f"Pressure 2 = {Bernoullis_equation.Bernoullis_equation_p2(p2)} Pascal")
```

#### Algebraically solving Bernoulli’s equation for P2 - Using functional programming

```python
@handcalc(jupyter_display = True)
def Bernoullis_equation_p2(p1:float,rho:float,v1:float,v2:float, g:float,h1:float,h2:float):
    p2 = (p1)+(1/2*rho*v1**2)+rho*g*h1 - (1/2*rho*v2**2)-(rho*g*h2)
    return p2
```

```python
p2 = Bernoullis_equation_p2(12300,1090,3,0.750,9.81, 0,1.20)
```
