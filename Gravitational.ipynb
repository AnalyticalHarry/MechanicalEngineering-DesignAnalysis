```python
import math
import pandas as pd
```

```python
%%time
force = [] #Newton
distance = [x for x in range(10,100)] #10 to 100 meter
mass1 = [x for x in range(800,1500)] # 800 to 1500 kg
mass2 = [x for x in range(800,1500)] #800 to 1500kg
G = 6.674 * (1/100000000000) #6.674×10-11 N m2/kg2
for i in distance:
    for j in mass1:
        for k in mass2:
            gravitational = G*((j*k)/(i*i))
            force.append(gravitational)
```

```python
dataset = {"Force (Newton)":force}
df_force = pd.DataFrame(dataset)
df_force
```
